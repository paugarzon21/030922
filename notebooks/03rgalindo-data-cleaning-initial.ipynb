```python
from pathlib import Path
import pandas as pd
import os
import numpy as np
```

```python
root_dir=Path('.').resolve().parent
file_name="llamadas123_julio_2022.csv"
file_path=os.path.join(root_dir, "data","row", file_name) 
data=pd.read_csv(file_path,sep=";", encoding="latin-1") 
data
```

```python
#Eliminar duplicados
print ('forma incial', data.shape)
data=data.drop_duplicates() # Encontrar filas duplicadas, desechar los duplicados y el resultado escribirlo en data
data.reset_index(inplace=True, drop=True) #definir indices de la nueva tabla, adicional inplace sobreescribe en el mismo lugar de memora
print ('forma modificada', data.shape)
```

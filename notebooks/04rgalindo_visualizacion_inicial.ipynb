# VISUALIZACION DE DATOS


```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#Datos numericos
#scatter plot (diagramas de dispersion)
x= np.arange(1,20)
y= x**2
y1=x**2 + 5*x
plt.scatter(x,y, s=100,color='pink', marker="*", label="x**2")
plt.scatter(x,y1, s=100,color='blue', marker="+", label="x**2+ 5")
plt.legend(title="Funcion")
plt.title('Funcion de x', fontsize= 20)
plt.ylabel("y")
#plt.xlim(0,5)
plt.grid()
plt.text(5,300,'este texto')
for val_x in x:
    plt.text(val_x, val_x**2, f'({val_x},{val_x**2})')
```
